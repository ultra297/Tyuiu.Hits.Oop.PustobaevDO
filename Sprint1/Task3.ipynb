```
class Person
{
    private string name;
    private int age;
    private string gender;

    public Person(string name, int age, string gender)
    {
        this.name = name;
        this.age = age;
        this.gender = gender;
    }

    public string Name
    {
        get { return name; }
        set { name = value; }
    }

     public int Age
    {
        get { return age; }
        set { age = value;}
    }

    public string Gender
    {
        get { return gender; }
        set { gender = value; }
    }

    public virtual string GetInfo()
    {
        return $"Это {Name}, возраст: {Age}, пол: {Gender}.";
    }   

    public virtual string SayHello()
    {
        return $"Привет, меня зовут {Name}.";
    }
    
}

class Student : Person
{
    private string university;
    private int course;

    public Student(string name, int age, string gender, string university, int course) : base(name, age, gender)
    {
        this.University = university;
        this.Course = course;
    } 
    
    public string University
    {
        get { return university; }
        set { university = value; }
    }

    public int Course
    {
        get { return course; }
        set { course = value; }
    }

    public override string SayHello()
    {
        return $"Привет! Я {Name}, мне {Age} лет, пол: {Gender}, я учусь в {University}, на {Course} курсе";
    }
}

class Employee : Person
{
    private string company;
    private decimal salary;

    public Employee(string name, int age, string gender, string company, decimal salary) : base(name, age, gender)
    {
        this.Company = company;
        this.Salary = salary;
    }

    public string Company
    {
        get { return company; }
        set { company = value; }
    }

    public decimal Salary
    {
        get { return salary; }
        set { salary = value; }
    }

    public override string GetInfo()
    {
        return $"Это {Name}, возраст: {Age}, пол: {Gender}, компания: {Company}, зарплата: {Salary}";
    }
}

class Teacher : Person
{
    private string subject;
    private int experience;

    public Teacher(string name, int age, string gender, string subject, int experience ) : base (name, age, gender)
    {
        this.Subject = subject;
        this.Experience = experience;
    }

    public string Subject
    {
        get { return subject; }
        set { subject = value; }
    }

    public int Experience
    {
        get { return experience; }
        set { experience = value; }
    }

    public override string SayHello()
    {
        return $"Привет! Я {Name}, мне {Age} лет, пол: {Gender}, я преподаю {Subject}, стаж работы: {Experience}";
    }

    public void MoveStudentToNextCourse(Student student)
    {
        if(student.Course >=6)
        {
            return;
        }
        student.Course++;
    }

    public void UpgradeEmployee(Employee employee)
    {
        employee.Salary *= 1.1m;
    }
}

var student = new Student("Dmitry", 19, "Man", "URFU", 2);
var teacher = new Teacher("Olga", 55, "Women", "math", 20 );
var employee = new Employee("Maria", 20, "Women", "Google", 43000);

Console.WriteLine(student.SayHello());

teacher.MoveStudentToNextCourse(student);

Console.WriteLine(student.SayHello());

Console.WriteLine(employee.GetInfo());

teacher.UpgradeEmployee(employee);

Console.WriteLine(employee.GetInfo());
```
